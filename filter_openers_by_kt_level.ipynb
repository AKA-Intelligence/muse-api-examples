```python
from client_util import is_easy
from tqdm import tqdm
```

```python
with open('self-dialogue-openers.txt','r') as fin:
    openers = fin.read().split('\n')
```

```python
lvl0 = []
lvl1 = []
lvl2 = []
lvl3 = []
```

```python
for opener in tqdm(openers):
    if is_easy(opener,0) and opener.endswith('?'):
        lvl0.append(opener) 
    if is_easy(opener,1) and opener.endswith('?'):
        lvl1.append(opener)
    if is_easy(opener,2) and opener.endswith('?'):
        lvl2.append(opener)
    if is_easy(opener,3) and opener.endswith('?'):
        lvl3.append(opener)
```
